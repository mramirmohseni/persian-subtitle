```python
import torch
from transformers import MBartForConditionalGeneration, MBartTokenizer
import gradio as gr
import requests
```

```python
def download_model_from_huggingface(hf_url, output_path):
    response = requests.get(hf_url)
    with open(output_path, 'wb') as f:
        f.write(response.content)
    return output_path

hf_url = "https://huggingface.co/alimansouri/mBart-fineTuned/resolve/main/final_model.pth"
```

```python
def load_finetuned_mbart():
    model_path = "final_model.pth"
    download_model_from_huggingface(hf_url, model_path)

    model = MBartForConditionalGeneration.from_pretrained("facebook/mbart-large-50-many-to-many-mmt")
    model.load_state_dict(torch.load(model_path))
    model.eval()
    return model
```

```python
device = 'cuda' if torch.cuda.is_available() else 'cpu'
```

```python
load_finetuned_mbart()
```

```python
model = load_finetuned_mbart()
```

```python
device = 'cuda' if torch.cuda.is_available() else 'cpu'
model.to(device)
```

```python
tokenizer = MBartTokenizer.from_pretrained("facebook/mbart-large-50-many-to-many-mmt")

tokenizer.src_lang = 'en_XX'
tokenizer.tgt_lang = 'fa_IR'

def translate(text):
    print(f"Input text: {text}")

    inputs = tokenizer(text, return_tensors='pt', max_length=512, truncation=True).to(device)

    with torch.no_grad():
        generated_tokens = model.generate(
            inputs['input_ids'],
            attention_mask=inputs['attention_mask'],
            max_length=512,
            num_beams=10,
            length_penalty=2.0,
            early_stopping=True
        )

    translation = tokenizer.decode(generated_tokens[0], skip_special_tokens=True)
    print(translation)
    return translation
```

```python
gr_interface = gr.Interface(
    fn=translate,
    inputs=gr.Textbox(label="Input Text", placeholder="Enter text to translate"),
    outputs="text",
    title="MBart Translation"
)
```

```python
gr_interface.launch()
```
