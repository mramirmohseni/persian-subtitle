```python
import torch
from transformers import MT5ForConditionalGeneration, MT5Tokenizer
import gradio as gr
from mega import Mega
```

```python
def download_model_from_mega(mega_url, output_path):
    mega = Mega()
    file = mega.download_url(mega_url, output_path)
    return output_path


mega_url = "https://mega.nz/file/5tx3kQ5Q#KXpPCQPJ4LRH_3T7cfvuuJQEEhQv8MlZe5lqIkKRbLc"
model_path = "./"
```

```python
def load_finetuned_mt5():
    download_model_from_mega(mega_url, model_path)

    model = MT5ForConditionalGeneration.from_pretrained("google/mt5-small")

    model.load_state_dict(torch.load(model_path))

    model.eval()
    return model
```

```python
load_finetuned_mt5()
```

```python
model_path = "/content/final_model.pth"

device = 'cuda' if torch.cuda.is_available() else 'cpu'

model = MT5ForConditionalGeneration.from_pretrained("google/mt5-small", low_cpu_mem_usage=True)

model.load_state_dict(torch.load(model_path))

model.eval()

tokenizer = MT5Tokenizer.from_pretrained("google/mt5-small")

model.to(device)
```

```python
def translate(text):
    print(f"Input text: {text}")

    inputs = tokenizer(text, return_tensors='pt', max_length=512, truncation=True).to(device)

    with torch.no_grad():
        generated_tokens = model.generate(
            inputs['input_ids'],
            attention_mask=inputs['attention_mask'],
            max_length=512,
            num_beams=10,
            length_penalty=2.0,
            early_stopping=True
        )

    translation = tokenizer.decode(generated_tokens[0], skip_special_tokens=True)
    print(translation)
    return translation
```

```python
gr_interface = gr.Interface(
    fn=translate,
    inputs=gr.Textbox(label="Input Text", placeholder="Enter text to translate"),
    outputs="text",
    title="mT5 Small Translation"
)
```

```python
gr_interface.launch()
```
